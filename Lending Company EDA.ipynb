```
# Creating the function to remove the percentage form interest rate column

def trim(string):
  return string.strip('%')

spark.udf.register('trimpercent',trim)
```

```
%sql

select int_rate,cast(trimpercent(int_rate) as float) as int_rate_float from loan_data_cleaned;
```

```
%sql

select bad_loan, cast(trimpercent(int_rate) as float) as int_rate from loan_data_cleaned;
```
